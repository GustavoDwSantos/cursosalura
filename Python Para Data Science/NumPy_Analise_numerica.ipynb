### Multiplicação

Nós chutamos um coeficiente angular, porem temos uma equação que calcula esse coeficiente

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
```

Coeficiente Linear

```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y)
```

### Motivo da regrassão

```python
plt.plot(datas,Moscow)
plt.plot(x,y)
```

```python
plt.plot(datas,Moscow)
plt.plot(x,y)
plt.plot(41.5,41.5*a+b,'*r')
```

### Reprodutibilidade de Numeros aleatorios

Quando geramos numeros aleatorios pelo computador, possuimos diversos calculos para gerar esses numeros, e eles partem sempre de uma semente 

Podemos definir essa semente para gerarmos sempre a mesma sequencia

```python
np.random.seed(16)
np.random.uniform(low=0.10,high=0.90,size=100)
```

Com isso podemos realizar o mesmo resultado diversas vezes, partindo da mesma semente

```python
np.random.seed(16)
coef_angulares = np.random.uniform(low=0.10,high=0.90,size=100)
norma2 = np.array([])
for i in range(100):
    norma2 = np.append(norma2, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
```

```python
norma2.min()
```

### Salvando os Dados

```python
norma2
```

```python
coef_angulares
```

Para salvarmos os dados dos arrays que criamos temos que primeiro estoca-los em colunas

```python
dados = np.column_stack([norma2, coef_angulares])
```
